Импорт необходимых библиотек

```python
%matplotlib inline

from sklearn import datasets 
from sklearn.datasets import load_iris

import pandas as pd
import numpy as np

import matplotlib.pyplot as plt
import seaborn as sns
import matplotlib as mpl
from matplotlib.colors import ListedColormap

from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
from sklearn import metrics
from sklearn.metrics import confusion_matrix, classification_report
from sklearn.neighbors import KNeighborsClassifier
```

Загрузка данных. Определение таргета

```python
iris = datasets.load_iris()
X = iris.data
y = iris.target
print (iris.feature_names)
print (iris.target_names)
```

Разбиение данных на обучающую и тестовую выборки в соотношении 80/20

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2)
```

Построение модели. Определение основных оценок модели таких как Accuracy/Precision/Recall. Подобрано минимальное число neighbors = 7 для получения Accuracy 0.97. А так же визуализация разделения на 3 класса

```python
knn = KNeighborsClassifier(n_neighbors=7, weights='distance') 
knn.fit(X_train, y_train)
pred = knn.predict(X_test)

print('Accuracy of KNN on test: {:.2f}'.format(metrics.accuracy_score(y_test,pred)))
print('Confusion matrix: \n{}\n'.format(confusion_matrix(y_test,pred)))
print('Classification report: \n{}\n'.format(classification_report(y_test,pred)))
```
